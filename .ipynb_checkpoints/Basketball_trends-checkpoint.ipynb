```python
def myplot(player_list):
    Col = {"KobeBryant":"black","JoeJohnson":"red","LeBronJames":"green","CarmeloAnthony":"purple","DwightHoward":"magenta","ChrisBosh":"brown","ChrisPaul":"cyan","KevinDurant":"gray","DerrickRose":"orange","DwayneWade":"yellow"}
    Mark = {"KobeBryant":"s","JoeJohnson":"o","LeBronJames":"P","CarmeloAnthony":"^","DwightHoward":"p","ChrisBosh":"+","ChrisPaul":"D","KevinDurant":"H","DerrickRose":"1","DwayneWade":"2"}
    for name in player_list:
        plt.plot(Games[Pdict[name]], c=Col[name], ls='--', marker=Mark[name], ms=7, label=name)
    plt.legend()
    plt.xticks(range(0,10), Seasons, rotation = 'vertical')
    plt.show()
    
myplot(["KobeBryant", "LeBronJames", "DerrickRose"])
```

```python
#Fixing the inputs
def mvp_nba(data, player_list=Players, badge=True):
    Col = {"KobeBryant":"black","JoeJohnson":"red","LeBronJames":"green","CarmeloAnthony":"purple","DwightHoward":"magenta","ChrisBosh":"brown","ChrisPaul":"cyan","KevinDurant":"gray","DerrickRose":"orange","DwayneWade":"yellow"}
    Mark = {"KobeBryant":"s","JoeJohnson":"o","LeBronJames":"P","CarmeloAnthony":"^","DwightHoward":"p","ChrisBosh":"+","ChrisPaul":"D","KevinDurant":"H","DerrickRose":"1","DwayneWade":"2"}
    for name in player_list:
        plt.plot(data[Pdict[name]], c=Col[name], ls='--', marker=Mark[name], ms=7, label=name)
    plt.legend()        
    plt.xticks(range(0,10), Seasons, rotation = 'vertical')
    plt.show()
```
